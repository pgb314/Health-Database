```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = df[["Age-standardized suicide rates (per 100 000 population)", "Interpersonal Violence"]]
x = df["Age-standardized suicide rates (per 100 000 population)"]
y = df["Interpersonal Violence"]
plt.scatter(x, y)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.regplot(x= df["year"], y = df["Age-standardized suicide rates (per 100 000 population)"])
plt.title('GDP per capita vs. Alzheimer')
```

```python
plt.figure(figsize=(14,8))
sns.regplot(x= df["per_capita-HIV/AIDS"], y = df["Adolescent birth rate (per 1000 women aged 15-19 years)"],scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.title('Adolescent birth rates  vs. GDP per capita')
```

```python
plt.figure(figsize=(14,8))
sns.regplot(x= df["year"], y = df["Total NCD Deaths (in thousands)"])
plt.title('Suicide Rates vs. Year')
```
